```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define Cobb-Douglas function
def f(K, L, alpha):
    return K**alpha * L**(1 - alpha)

# Parameters
alpha = 0.3  # Capital share of income 
K = np.linspace(1, 100, 100)  # Capital stock
L = np.linspace(1, 100, 100)  # Labor input 
K, L = np.meshgrid(K, L)  # Create a meshgrid for K and L

Y = f(K,L, alpha = 0.25)

```

```python
# Erstelle den 3D-Plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Zeichne die Oberflche und fge eine Colorbar hinzu
surf = ax.plot_surface(K, L, Y, cmap='viridis', alpha=0.7, edgecolor='none')
cbar = fig.colorbar(surf, shrink=0.5, aspect=5)  # Die Colorbar zeigt die Werte von Y
cbar.set_label('Output (Y)')  # Beschriftung der Colorbar

# Achsenbeschriftungen und Titel
ax.set_xlabel('Capital (K)')
ax.set_ylabel('Labour (L)')
ax.set_zlabel('Output (Y)')
ax.set_title('Cobb-Douglas Production Function')

plt.show()


```
